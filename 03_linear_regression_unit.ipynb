# 선형회귀 (Dense Layer)

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

# 데이터 준비

```python
X_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24.])
```

### 모델

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(units=1, input_shape=(1,)),        #완전 연결층(Fully Connected Layer) , layer 1개, unit 1개, 입력 1개
])
```

```python
model.summary()
```

```python
## 컴파일

# adam = tf.keras.optimizers.Adam(learning_rate=0.01) learning rate 변경 / optimizer = adam
# sgd = tf.keras.optimizers.experimental.SGD()
model.compile(
    optimizer = 'sgd',
    loss = 'mse',
    metrics = ['mse', 'mae']
)

```

```python
## 학습

model.fit(X_train, y_train, 
          epochs = 100, 
          batch_size=len(X_train), 
          verbose = 1)
```

```python
# 학습된 파라미터

model.get_weights()
```

```python
# 평가

model.evaluate(X_train, y_train)
```

```python
# 예측

X_test = [10.]

y_pred = model.predict(X_test)
y_pred
```
