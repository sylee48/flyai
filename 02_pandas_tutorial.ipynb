```python
sr = pd.Series([1, 2, 3, 4], name = 'A')
               
print('type : ', type(sr))
print('dtype : ', sr.dtype)
print(sr)
```

```python
sr = pd.Series([1, 2, 3, 4], index = ['A', 'B', 'C', 'D'])
               
print('type : ', type(sr))
print('dtype : ', sr.dtype)
print(sr)
```

```python
x = [['Jackson', 68, True, 1.5, 'A'],
     ['Liam', 74, True, 1.1, 'A'],
     ['Emma', 20, False, 0.89, 'C']]
df = pd.DataFrame(x,
            columns = ['Name', 'Weight', 'Option', 'Etc', 'Type'])
print('type : ', type(df))
print('shape : ', df.shape)
print(df)
```

```python
data = { 
    'Name':['Jackson', 'Emma', 'Noah', 'James'],
    'Weight':[68, 74, 77, 78],
    'Option':[True, True, False, False],
    'Rate':[0.21, 1.1, 0.89, 0.91],
    'Type':['A','A','C','C']
}
df = pd.DataFrame(data)

print('type : ', type(df))
print('shape : ', df.shape)
print(df)
```
