```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.__version__
```

### 기본 그래프

```python
x = [1, 3, 5]
y = [2, 8, 11]
```

```python
plt.plot(x, y)
plt.title('Graph')
plt.xlabel('X', loc='right')
plt.ylabel('Y')
```

```python
plt.rcParams.get('figure.figsize')
```

### Figure, Axes

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(x,y)
ax1.set_title('Graph')
ax4.plot(x,y)
plt.show()
```

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, label='Eng', marker='^', linewidth=3, linestyle='-.')
plt.plot(x, z, label='Math', color='#7FFFD4', linewidth=3, linestyle='-.')
plt.legend(loc='lower right')
```

### 막대그래프 barplot

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 2, 5, 4, 6]

plt.bar(x, y, color='r', width=0.1)
plt.title('Graph')
plt.xlabel('X', loc='right')
plt.ylabel('Y')
```

### 히스토그램 (histogram)

```python
import numpy as np

mydata = np.random.randn(500)
plt.hist(mydata, color='pink', bins=50)
plt.show()
```

```python
img1 = Image.open('닥스훈트.jpg')
img2 = Image.open('포메.jpg')
img3 = Image.open('코기.jpg')
img4 = Image.open('리트리버.jpg')

images = [img1, img2, img3, img4]

fig, ax = plt.subplots(2, 2)

ax = ax.reshape(-1)

for i in range(4):
    ax[i].imshow(images[i])
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title('Dog')
    
plt.show()
```
