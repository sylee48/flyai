```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.__version__
```

```python
df = sns.load_dataset('penguins')
df
```

### Axes Level

```python
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.histplot(data=df, x='species', hue='species')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax=axes[0, 0])
sns.histplot(data=df, x='flipper_length_mm', hue='species', ax=axes[1, 0])
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax=axes[1, 1])
sns.histplot(data=df, x='flipper_length_mm', hue='species', ax=axes[0, 1])

plt.show
```
