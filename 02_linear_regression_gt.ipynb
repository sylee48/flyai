# 선형회귀 )Gradient Tape.)
--- 
- GradientTape : 텐서플로우의 자동미분 기능

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

### 데이터 준비

```python
X_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24.])
```

### 경사하강법

```python
# 초기화
W = tf.Variable(0.0)
b = tf.Variable(0.0)

y_hat = W * X_train + b
```

```python
# MSE
cost = tf.reduce_mean(tf.square(y_train - y_hat))
```

```python
# 자동미분 기능
lr = 0.01

with tf.GradientTape() as tape:
     y_hat = W * X_train + b
     cost = tf.reduce_mean(tf.square(y_train - y_hat))

W_grad, b_grad = tape.gradient(cost, [W, b])

# W = W - lr * W_grad
W.assign_sub(lr * W_grad)
b.assign_sub(lr * b_grad)
```

```python
W.numpy(), b.numpy()
```

### 전체 코드

```python
# 초기화
W = tf.Variable(0.0)
b = tf.Variable(0.0)

epochs = 5000
lr = 0.01

for i in range(epochs):
    with tf.GradientTape() as tape:
        y_hat = W * X_train + b
        cost = tf.reduce_mean(tf.square(y_train - y_hat))

        W_grad, b_grad = tape.gradient(cost, [W, b])

        # W = W - lr * W_grad
        W.assign_sub(lr * W_grad)
        b.assign_sub(lr * b_grad)

        if i % 200 == 0:
            print(f'Epoch:{i}, cost:{cost.numpy():10f}, W:{W.numpy():10f}, b:{b.numpy():10f}')
```
